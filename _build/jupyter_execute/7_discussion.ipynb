# Discussion

## Microplastiques

Sur l’ensemble des 25 plages étudiées, les résultats obtenus montrent une abondance moyenne en microplastiques de 16’500 ± 15’600 par m². Ce chiffre, issue des comptabilisations effectuées à la loupe binoculaires, peut être pondéré en se basant sur une étude réalisée en Méditerranée. En effet, Constant et al, ([Constant et all., 2019](constant_et_all)), ont déterminé des pourcentages d’identification des différentes particules. Les fibres y sont identifiées comme étant bien du plastique dans 37% des cas, les fragmentles fragments à 50%, les mousses à 81% et les films à 67%. En appliquant cette pondération aux données du projet Pla’stock, la contamination estimée en plastique serait de 7’200 ± 6’800 particules par m².

Avec le tableau 7.1 ci-dessous, il est visible que les abondances mesurées dans le cadre de l’étude Pla’stock (16’500 ± 15’600 particules/m²) sont largement supérieures à la moyenne mesurée en 2014 sur le Léman et les autres lacs Suisses. Ceci confirme les observations faites par Frei et all, ([Frei et all., 2021](frei_et_all)), sur le lac de Lugano. Il aurait été intéressant d’avoir un état des lieux en 2014 de la pollution plastique sur les plages du lac de Lugano afin de comparer l’évolution entre les deux lacs.

La comparaison avec une étude en Méditerranée, sur la plage du Golf du Lion, est encore plus flagrant ([Constant et all., 2019](constant_et_all)). Contrairement aux autres études Frei et all, Constant et all et ([Faure et all., 2018](faure_et_all)), les particules comptabilisées dans le cadre de la présente étude n’ont pas été caractérisées, la valeur pondérée permet d’obtenir une abondance probable de particules plastique et une meilleure comparaison avec les autres études. La valeur non pondérée constitue la valeur maximale de particules dans les échantillons.

Les méthodes de prélèvements sont similaires dans les différentes études. En revanche la taille des échantillons varie. Constant et all et Faure et all, ont fait des carottages de 0,5 x 0,5 m contre 0,1 x 0,1 m pour l’étude de Frei et all et l’étude Pla’stock. Cette nouvelle taille semble devenir la norme pour les échantillonnages de microplastiques sur les plages.

## Macroplastiques

Les résultats de Pla’stock suggèrent que les quantités de certains déchets sauvages sont en baisse par rapport aux sessions d’échantillonnage précédentes sur les plages du Léman (Tableau 7.2). Le projet IQAASL est la campagne d’échantillonnages qui est la plus comparable avec l’étude Pla’stock. Les différences sont la supervision des bénévoles, le nombre de passages et le nombre de plages suivies. Pour l’étude Pla’stock, les responsables de secteur ont reçu une formation en salle pour réaliser les inventaires alors que dans le cadre d’IQAASL, les bénévoles étaient accompagnés sur les plages. Les autres projets mentionnés ne sont pas des études à large échelle. Il s’agit plutôt de collectes spontanées par des groupes sur des nombres de secteurs restreints. Pla’stock est la première étude à avoir analysés tant des plages suisses que françaises.

Dans l’ensemble, l’étude fait état d’une valeur médiane de 2 pièces/mètre contre 3,7 à 4,4 pièces/mètre dans le passé. Cette réduction est conforme aux conclusions de 2020 -– 2021 combinant les résultat du suivi par SLR et de l’étude IQAASL ([OFEV](OFEV)).

## Stock de plastiques sur les plages du Léman

 Pour extrapoler les résultats de notre étude à l’ensemble des plages du Léman, la superficie des plages publiques a été estimée sur la base d’orthophotos et en se référant à la liste des plages de la CIPEL. Cette liste a été adaptée, car certaines plages sont uniquement des espaces de baignade dépourvues de grèves “naturelles” (enrochements, dalles en pierres ou en beton). De plus, trois plages faisant parti lde l’étude Pla’stock ont été ajoutées à la liste des plages. Il s’agit d’une plage semi-privée à Crans (VD), de la plage de la Pichette sur la commune de Corseaux et de la plage de l’Empereur à Noville (Grangettes). Au total, la superficie des 90 plages retenues correspond à 68’399 m² pour un linéaire de 12’688 m.
 
Selon cette estimation, la contamination plastique s’élève à 492'000’000 ± 465'000'000 particules sur les plages du Léman en se basant sur la moyenne pondérée et 1’129'000'000 ± 1'067'000'000 sans pondérationon. 
Pour les macroplastiques, sur les 12’688 m de plages publiques, en extrapolant la moyenne de 3,42 pièces par mètre recensé dans le cadre du projet Pla’stock, il y aurait 440'092 morceaux de plastiques sur les plages publiques.
